```python
import openai
from openai import OpenAI

#set your api
Client=OpenAI(api_key='')

def get_recommendation(season , budget , continent):
  PROMPT=f"""
  Recommend three travel destination for :
  season : {season}
  budget : {budget}
  continent : {continent}

  Respond in JSON format fields: Destination ,Region ,Estimated_Cost.
  """
  try :
    response=Client.chat.completions.create(
       model="gpt-3.5-turbo",
       messages=[
           {"role":"user","content":PROMPT}
       ]
       )
    return response.choices[0].message.content
  except Exception as e:
    return f"Error: {str(e)}"

#interactive Tavel Assistant
season = input("Preferred Season,Eg:Summer or Winter ")
continent=input("Enter the continent: Eg-Asia,Europe")
budget=input("Enter Budget , Eg:low ,high ,medium ")
result=get_recommendation(season,continent,budget)
print(result)
```
