```python
import openai

openai.api_key = ""  # üîê Replace with your key
```

```python
def chat_with_gpt(prompt):
    client = openai.OpenAI(api_key=openai.api_key)  # Initialize the client with the API key

    response = client.chat.completions.create(
        model="gpt-3.5-turbo",  # or "gpt-4" if available
        messages=[
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": prompt}
        ]
    )
    return response.choices[0].message.content

# Example usage
user_input = "Explain what is machine learning in simple terms."
output = chat_with_gpt(user_input)
print("üîç GPT Response:", output)
```

```python
import ipywidgets as widgets
from IPython.display import display

input_box = widgets.Textarea(placeholder='Ask me anything...')
output_box = widgets.Output()

def on_submit(change):
    output_box.clear_output()
    with output_box:
        result = chat_with_gpt(change['new'])
        print(result)

input_box.observe(on_submit, names='value')

display(input_box, output_box)
```
