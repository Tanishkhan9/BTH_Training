```python
'''
#Prompt Engineering : Skill of writing effective inputs or prompts to guide LLMs(AI models), (For ex = like Chatgpt, Gemini , Llama , Mistral)  to  get accurate and helpful result.

#Bad Prompts : "Tell me about Ai"

#Good Prompt  : "Explain Artificial Intelligence to a 10 year old , with example"

#Principle of Good Prompts:
                            *Clearity -(Use simple and clear instruction)
                              Ex= Summarize the paragraph in 3 bullet points
                            *Context  -(Provide necessary background)
                              Ex= Given this resume write a cover letter for data scientist
                            *Specify  -(Clearly define the tasks)
                              Ex= Write 4 line about the moon
                            *Role Playing -(Ask the model to specific character)
                              Ex= You are a maths teacher . Explain binary numbers.
                            *Iteration  -(Refine prompts based on results)
                              Ex= Improve this prompt to generate more creative ideas.

#Prompting Techniques:
  1.Zero Shot  -No example given
    Ex= Transfer this sentence in hindi
  2.Few Shot   -Provide few example
    Ex= Translate :cat = chat
  3.Chain of Thought    -Ask for step by step reasoning
    Ex= Explain your steps for solving =====  (24/4)*2

'''
```

```python
from fpdf import FPDF
#INFO
YOUR_NAME= "Tanish khan"
#GITHUB
GITHUB_URL="https://github.com/Tanishkhan9"
LINKEDIN_URL="https://www.linkedin.com/in/tanish-k-4b966b255/"
PROFILE_IMAGE="/content/PIC.jpg"  #Replace this with your uploaded file name
class PDF(FPDF):
  def header(self):

    if self.page_no() == 1:
      return
    self.set_font("Arial","B",12)
    self.cell( 0 , 10 , " GeneratingAi and LLMs",ln=True,align="C")
    self.ln(5)
  def footer(self):
      self.set_y(-15)
      self.set_font("Arial","I",8)
      self.cell(0,10,f'CREATED BY{YOUR_NAME}|GitHub:{GITHUB_URL}|Linkedin:{LINKEDIN_URL}',align="C")
  def title_page(self):
    if PROFILE_IMAGE:
      self.image(PROFILE_IMAGE,x=80,y=30,w=50)
      self.ln(80)
    self.set_font("Arial","B",12)
    self.set_y(80)
    self.cell(0,10,"GenrerativeAI and LLMs  ",ln=True ,align="C")
    self.set_font("Arial","",14)
    self.ln(10)
    self.cell(0, 10,f"created by :{YOUR_NAME}",ln=True,align="C")
    self.set_font("Arial","",12)
    self.cell(0,10,f"GitHub:{GITHUB_URL}",ln=True,align="C")
    self.cell(0,10,f"Linkedin:{LINKEDIN_URL}",ln=True,align="C")

  def section_body(self,text):
    self.set_font("Arial","",11)
    self.multi_cell(0,8,text)

  def section_title(self,title):
    self.set_font("Arial","B",14)
    self.set_text_color(0,51,102)
    self.cell(0,10,title,ln=True)
    self.set_text_color(0,0,0)
    self.ln(2)
  def section_body(self,text):
    self.set_font("Arial","",11)
    self.multi_cell(0,8,text)
    self.ln()
  def Capabilities_table(self):
    self.set_font("Arial","B",12)
    self.cell(60,10,"Task",1,0,'C')
    self.cell(120,10,"Example Prompt",1,1,'C')
    self.set_font("Arial","",11)
    rows=[
        ("Summarization","Summarize the article about AI."),
        ("Translation","Translate 'Good morning 'to Spanish."),
        ("Q&A ","What is quantum computing"),
        ("Text Generation","Write a story about a robot artist ."),
        ("Code Generation","Write a python function for factorial."),
        ("Role Play","Act as a Paris tour goude"),

          ]
    for task, prompt in rows:
      self.cell(60,10,task,1)
      self.cell(120,10,prompt,1,1)
#Create PDF
pdf=PDF()
pdf.add_page()
pdf.title_page()
pdf.add_page()


#Section
pdf.section_title("1. Overview of Generative AI")
pdf.section_body(
    "Generative AI creates new content instead of just analyzing existing data.\n\n"
    "It can generate:\n"
    "- Text (e.g., essays, summaries)\n"
    "- Images (e.g., DALLÂ·E)\n"
    "- Code (e.g., GitHub Copilot)\n"
    "- Audio and Music\n"
    "- Videos (e.g., Sora, Runway)"
)

pdf.section_title("2. What Are LLMs?")
pdf.section_body(
    "Large Language Models (LLMs) are deep learning models trained on vast amounts of text data.\n\n"
    "Capabilities include:\n"
    "- Understanding context\n"
    "- Generating human-like text\n"
    "- Answering questions\n"
    "- Translating languages\n"
    "- Summarizing content\n\n"
    "Popular LLMs: GPT-4, Claude, Gemini, LLaMA"
)

pdf.section_title("3. How LLMs Work")
pdf.section_body(
    "LLMs are trained to predict the next word in a sentence using a technique called 'next-word prediction.'\n\n"
    "The architecture used is called a Transformer, which uses self-attention to understand context.\n"
    "Training involves vast text corpora and multiple stages of fine-tuning."
)

pdf.section_title("4. Exploring LLM Capabilities")
pdf.section_body("Here are examples of what LLMs can do with well-crafted prompts:")
pdf.Capabilities_table()
pdf.ln()
pdf.set_font("Arial", "I", 10)
pdf.cell(0, 10, "Tip: The clearer your prompt, the better the output!", ln=True)


#Save
pdf.output("genAI.pdf")
```
