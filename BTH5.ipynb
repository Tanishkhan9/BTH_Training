```python
'''
Neutral Network: A computational model  inspired by the human brain

Basic structures:  #Neurons - Unit that take input and produce outputs

                    Layers - 3 layers
                              #1-Input Layer = Receives raw data
                              #2-Hidden Layer = This is where computation happens
                              #3-Output Layer = Produce final output

                    Connections - Data flow between layers through weighted connectons
                                  (this weight are learn during training model)

Activation Functions: They helps the network to learn complex pattern and complex data like cluster, curves ,irregular shapes not just straight line

                      1.Relu()    : Rectifier linear unit (fast ,efficient)=Max (0,X)
                      2.Sigmoid() :  Input varies between (0,1)    Good for probability
                      3.Tanh()    : like sigmoid but varies between (-1,1)    Better for gradients


Why non-linearity is important ?
1. Without activation function ,Neural network would just be linear transformation , which serverly limits what they can learn


What is Deep Learning:: DL refers to neural networks with multiple layers and enables model to learn hierarchical representation of data. (it works on hidden layers)


poiuyt0987654
SGD:stochastic gradient descent =use to solve complex data
                                =it is an optimization algorithm used to  machine learning to train model
 '''
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD

#XOR-dataset
x=np.array(
    [
        [0,0],
         [0,1],
          [1,0],
           [1,1]]
)
y=np.array([
    [0],
     [1],
      [1],
       [0]
    ])
#Model Creation

model=Sequential()
model.add(Dense(units=2,activation='sigmoid',input_dim=2))
model.add(Dense(units=1,activation='tanh'))
model.add(Dense(units=1,activation='sigmoid'))
model.compile(optimizer=SGD(learning_rate=0.6),loss='binary_crossentropy',metrics=['accuracy'])

#Train Model
model.fit(x, y, epochs=1000, verbose=0)


# predict prediction
print("Prediction: ",model.predict(x).round())
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

#And gate data set
x=np.array([
    [0,0],[0,1],[1,0],[1,1]
])
y=np.array(
    [0,0,0,1]
    )

#model building
model=Sequential()
model.add(Dense(units=2,activation='relu',input_dim=2)) #hidden layer
model.add(Dense(units=1,activation='sigmoid'))#output layer


#compile model
model.compile(loss='binary_crossentropy', optimizer='adam',metrics=['accuracy'])


#train model
model.fit(x,y,epochs=99,verbose=0)

#predictions
prediction=model.predict(x).round()
print("Andgate Prediction:",prediction.flatten())
```
